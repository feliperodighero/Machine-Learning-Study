```python
iris = datasets.load_iris()

clust_1 = scale(iris.data)
clust_2 = scale(np.random.rand(150, 4))
```

```python
inertia = []
for i in range(1, 8):
    kmeans = KMeans(n_clusters=i, n_init='auto')
    kmeans.fit(clust_1)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 8), inertia)
plt.title('Elbow Method')
plt.show()
```

```python
inertia = []
for i in range(1, 8):
    kmeans = KMeans(n_clusters=i, n_init='auto')
    kmeans.fit(clust_2)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 8), inertia)
plt.title('Elbow Method')
plt.show()
```
